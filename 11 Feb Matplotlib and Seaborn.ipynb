# Scatterplot

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
## plt is a page, figure means we will make a diagram, figsize is figure size

plt.figure(figsize = (6,2))
plt.scatter(x,y, c = 'r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatterplot')
plt.show()
```

```python
## controlling the intensity/transparency of the plot

plt.figure(figsize = (6,2))
plt.scatter(x,y, c = 'r', alpha= 0.5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatterplot')
plt.show()
```

```python
## changing the dot pattern into other

plt.figure(figsize = (6,2))
plt.scatter(x,y, c = 'r', alpha= 0.5, marker = 'x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatterplot')
plt.show()
```

```python
## changing the dot pattern into other

plt.figure(figsize = (6,2))
plt.scatter(x,y, c = 'r', alpha= 0.5, marker = '*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatterplot')
plt.show()
```

#### pie chart

```python
lang = ['C', 'python', 'java']
students = [20,100,40]
```

```python
plt.pie(students, labels= lang )
plt.title('Pie Chart')
plt.show()
```

```python
plt.pie(students, labels= lang, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
my_explode = (0.0,1.0,0)

plt.pie(students, labels= lang, autopct='%1.1f%%', explode= my_explode)
plt.title('Pie Chart')
plt.show()
```

```python
my_explode = (0.0,0.1,0.1)

plt.pie(students, labels= lang, autopct='%1.1f%%', explode= my_explode, shadow= True)
plt.title('Pie Chart')
plt.show()
```
